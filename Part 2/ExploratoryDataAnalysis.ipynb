```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
```

### Data Distribution:
We plot the label distribution of the train, test and val datasets. There are three labels: normal, bacterial pneumonia and viral pneumonia.

```python
train_path = "archive/chest_xray/train"
test_path = "archive/chest_xray/test"
val_path = "archive/chest_xray/val"

# Get the number of samples for each category and each dataset (train, test, val)
n_normal_train = len(os.listdir(os.path.join(train_path, "NORMAL")))
n_bacteria_train = len([f for f in os.listdir(os.path.join(train_path, "PNEUMONIA")) if "bacteria" in f])
n_virus_train = len([f for f in os.listdir(os.path.join(train_path, "PNEUMONIA")) if "virus" in f])

n_normal_test = len(os.listdir(os.path.join(test_path, "NORMAL")))
n_bacteria_test = len([f for f in os.listdir(os.path.join(test_path, "PNEUMONIA")) if "bacteria" in f])
n_virus_test = len([f for f in os.listdir(os.path.join(test_path, "PNEUMONIA")) if "virus" in f])

n_normal_val = len(os.listdir(os.path.join(val_path, "NORMAL")))
n_bacteria_val= len([f for f in os.listdir(os.path.join(val_path, "PNEUMONIA")) if "bacteria" in f])
n_virus_val = len([f for f in os.listdir(os.path.join(val_path, "PNEUMONIA")) if "virus" in f])

fig, ax = plt.subplots(1, 3, figsize=(20, 5))
ax[0].bar(["Normal", "Bacterial Pneumonia", "Viral Pneumonia"], [n_normal_train, n_bacteria_train, n_virus_train])
ax[0].set_title("Train Dataset Distribution")
ax[1].bar(["Normal", "Bacterial Pneumonia", "Viral Pneumonia"], [n_normal_test, n_bacteria_test, n_virus_test])
ax[1].set_title("Test Dataset Distribution")
ax[2].bar(["Normal", "Bacterial Pneumonia", "Viral Pneumonia"], [n_normal_val, n_bacteria_val, n_virus_val])
ax[2].set_title("Val Dataset Distribution")
plt.show()
```

### Example Images:
We plot three images of each category (normal, bacterial pneumonia, viral pneumonia)

```python
normal_images = [Image.open(os.path.join(train_path, "NORMAL", img)) for img in os.listdir(os.path.join(train_path, "NORMAL"))[:5]]
bacteria_images = [Image.open(os.path.join(train_path, "PNEUMONIA", img)) for img in[f for f in os.listdir(os.path.join(train_path, "PNEUMONIA")) if "bacteria" in f][:5]]
virus_images = [Image.open(os.path.join(train_path, "PNEUMONIA", img)) for img in [f for f in os.listdir(os.path.join(train_path, "PNEUMONIA")) if "virus" in f][:5]]

fig, ax = plt.subplots(3, 5, figsize=(20, 10))
for i, img in enumerate(normal_images):
    ax[0, i].imshow(img, cmap="gray")
for i, img in enumerate(bacteria_images):
    ax[1, i].imshow(img, cmap="gray")
for i, img in enumerate(virus_images):
    ax[2, i].imshow(img, cmap="gray")
for _ax, col in zip(ax[:, 0], ["Normal", "Bacterial Pneumonia", "Viral Pneumonia"]):
    _ax.set_ylabel(col, fontsize="x-large")
fig.tight_layout()
plt.show() 
```
